# What is the C++ programming language?

C++ is a low-level general-purpose programming language created as an extension to the C programming language. What does this mean? The language's **low-level** nature enables you, as the developer, to control all aspects of your application. With C++, you can instruct the computer to do almost anything, which is why it is such a powerful language. **general-purpose** means you can use C++ in a variety of scenarios. C++ is used to create other languages such as Java and Python, to create operating systems such as Andriod and Windows, and to develop applications such as Photoshop and Microsoft office. You've picked a great language to work with.

## Printing Text

Let's learn how to print text out on the console. The code cell below contains
code that prints the text "Hello C++ developer" on the console. We will break
down each component of the code.

```c++
#include <iostream>
```
This is known as an import statement. This instructs the computer to fetch additional code we will use in our application. We are importing additional code to work with IO streams in this case. We will use an IO Stream to display text in the console and accept user input. 

```c++
std::cout
```
This is known as a stream. The text sent to this stream will be displayed in the console. 

```c++
<<
```
This is known as a *stream operator* in this context. The *stream operator* takes the string on the right and sends it to the `std::cout` stream on the left. 

```c++
"hello world"
```
This text is known as a string. This is a string because the text is placed within double quotes`"`. We are writing this string to the console.

```c++
<< std::endl;
```
Finally, we end the line with `std::endl` tells the stream that the text we are sending to it is complete and put any additional text sent to it after it on a new line. Don't forget to end your C++ code lines with a colon `;`, so the compiler knows when the code ends.

To run the code cell below, press the play button to the left of the cell. Hover your mouse over the code cell, for the play button to appear.

````c++
#include <iostream>

std::cout << "Hello C++ developer" << std::endl;
````

## Accepting user input

In this section, we will learn how to accept user input from the keyboard. 

```c++
#include <string>
```
This import statement enables us to work with Strings in our code. 

```c++
using namespace std;
```
This changes the application's namespace to the standard library. Now we don't have to add `std::` in front of all resources from the C++ standard library. Example `std::cout` is now `cout`.

```c++
string input;
```

Here we are creating a variable. A variable is a bucket that holds a specific type of data. In this case, the variable's name is `input` and contains a string.

```c++
cout << "Please type something: "<< endl;
```

We print the message "Please type something: " in the console.

```c++
cin >> input;
```
This is where we accept user input. This will wait for the user to type something in and press enter. Then it will take the text the user typed in and put it in the variable named `input`. 

```c++
cout << "You typed: " << input << endl;
```
Finally, we print out what the user typed in with this line.

When you run the code cell below, you must type in your user input in the prompt above 👆.

````c++
#include <iostream>
#include <string>

using namespace std;

string input;

cout << "Please type something: "<< endl;
cin >> input;
cout << "You typed: " << input << endl;
````
